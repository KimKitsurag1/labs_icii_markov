```python
import pandas as pd
import numpy as np
```

1. Считываем CSV-файл

```python
df = pd.read_csv('lab2_data.csv')
```

```python
df.head()
```

2. Ищем список уникальных наименований документов

```python
x = df['Наименования документов, требованиям которых соответствует средство'].str.replace('Соответствует требованиям документов: ','')
x = x.str.split(', ', expand = True) # делим все строки запятыми
x = x.values.flatten() # делаем массив одномерным
x = x[~pd.isna(x)] # сбрасываем NaN значения
x = np.unique(x)
x
```

3. Выводим список средств соответствующие двум нормативным документам(в таких нормативные документы разделяет всего одна запятая)

```python
x = df['Наименования документов, требованиям которых соответствует средство'].str.replace('Соответствует требованиям документов: ','')
df['Наименование\r\nсредства (шифр)'][x.str.count(',')==1]
```

4. Выводим список средств с просроченными сертификатами

```python
d1 = pd.to_datetime(df['Срок\r\nдействия\r\nсертификата'],errors = 'coerce', dayfirst = True) < pd.to_datetime('26.12.2022')
df['Наименование\r\nсредства (шифр)'][d1]
```

5. Выводим список средств с просроченными сертификатами, но обязанными продолжать поддержку

```python
d2 = d1 & (pd.to_datetime(df['Информация об окончании срока технической поддержки, полученная от заявителя'],errors = 'coerce', dayfirst = True) > pd.to_datetime('26.12.2022'))
df['Наименование\r\nсредства (шифр)'][d2]
```

6. Выводим список средств прошедших испытания в частных организациях

```python
d3 = df['Испытательная\r\nлаборатория'].str.contains('ООО') 
d4 = df['Испытательная\r\nлаборатория'].str.contains('ЗАО') 
d5 = df['Испытательная\r\nлаборатория'].str.contains('АО') 
d6 = df['Испытательная\r\nлаборатория'].str.contains('ПАО') 
df[d3 | d4 | d5 | d6]['Наименование\r\nсредства (шифр)']
```
